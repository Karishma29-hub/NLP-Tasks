```python
import nltk
import string
from collections import Counter
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

articles = [
    "Artificial intelligence is transforming healthcare and education systems",
    "AI and machine learning are the future of technology and innovation",
    "Healthcare systems are improving with artificial intelligence solutions",
    "Education technology is growing rapidly with AI based tools"
]

# Convert all text to lowercase
articles = [article.lower() for article in articles]

# Remove punctuation, stopwords and tokenize
stop_words = set(stopwords.words('english'))
all_words = []

for article in articles:
    # remove punctuation
    for p in string.punctuation:
        article = article.replace(p, "")
    
    # tokenization
    words = word_tokenize(article)
    
    # remove stopwords
    filtered_words = [word for word in words if word not in stop_words]
    
    all_words.extend(filtered_words)

# Compute word frequency distribution
freq = Counter(all_words)

# Build vocabulary
vocabulary = set(all_words)

# Top 5 most frequent words
top5 = freq.most_common(5)

# Calculate Type-Token Ratio (TTR)
total_tokens = len(all_words)
unique_tokens = len(vocabulary)
ttr = unique_tokens / total_tokens

# Identify dominant topic based on frequency
# We can assume the most frequent meaningful word/topic is dominant
dominant_topic = top5[0][0]

# Output
print("Tokens after preprocessing:")
print(all_words)

print("\nWord Frequency Distribution:")
print(freq)

print("\nVocabulary (Unique Words):")
print(vocabulary)

print("\nTop 5 Most Frequent Words:")
for word, count in top5:
    print(word, ":", count)

print("\nTotal Tokens:", total_tokens)
print("Unique Tokens:", unique_tokens)
print("Type Token Ratio (TTR):", round(ttr, 2))

print("\nMost Dominant Topic:", dominant_topic)

# Visualize word frequency using bar chart
words = list(freq.keys())
counts = list(freq.values())

plt.figure(figsize=(10, 5))
plt.bar(words, counts, color="skyblue")
plt.title("Word Frequency in Technology News Articles")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
